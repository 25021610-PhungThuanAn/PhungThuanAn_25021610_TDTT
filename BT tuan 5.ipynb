```python
#W5A1
def Max_Of_Two(a, b):
    if a > b:
        return a
    else:
        return b
```

```python
#W5A2
def swap(a, b):
    a,b = b, a
    return (a, b)
```

```python
#W5A3
def snt_check(x):
    from math import isqrt
    if x == 2:
        return True
    elif x > 2:
        for i in range(2, isqrt(x) + 1):
            if x % i == 0:
                return False
        else:
            return True
    else:
        return False
```

```python
#W5A4
def so_hoan_hao_check(x):
    if x <= 0:
        return False
    tong = 0
    for i in range(1, x // 2 + 1):
        if x % i == 0:
            tong += i
    return tong == x
```

```python
#W5A5
def tim(danh_sach, k):
    n = len(danh_sach)
    for i in range(n):
        x = danh_sach[i]
        if x == k:
            return i
    return -1
```

```python
#W5A6
def tinh_giai_thua(n):
    if n < 0:
        raise ValueError
    tich = 1
    for i in range(1, n + 1):
        tich *= i
    return tich
```

```python
#W5A7
def tinh(num1, num2, operat):
    if operat == "+":
        ketqua = num1 + num2
    elif operat == "-":
        ketqua = num1 - num2
    elif operat == "*":
        ketqua = num1 * num2
    elif operat == "/":
        if num2 == 0:
            raise ValueError
    else:
        raise ValueError
        ketqua = num1 / num2
    return round(ketqua, 2)
```

```python
#W5A8
def Hamming_dis(x, y):
    z = x ^ y ## Toán tư XOR, nếu 2 bit giống nhau trả về 0, khác nhau trả về 1
    n = bin(z).count("1") # bin biến 1 dãy bit thành xâu, count(a): đếm số a trong str
    return n

def Hamming_dis(x, y):
    z = x ^ y
    count = 0
    temp = z
    while temp > 0:
        if temp % 2 == 1:
            count += 1
        temp //= 2
    return count
```

```python
#W5A9
def tong_chu_so(n):
    temp = n
    tong = 0
    while temp > 0:
        so = temp % 10
        tong += so
        temp //= 10
    return tong
```

```python
#W5A10
def dang_cau(a, b):
    so_sanh = {}
    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]
        if char_a in so_sanh:
            if so_sanh[char_a] != char_b:
                return False
        else:
            if char_b in so_sanh.values():
                return False
            else:
                so_sanh[char_a] = char_b
    return True
```
