```python
#W6A5
lst = list(map(int, input("Nhap vao cac so nguyen cach nhau boi 1 dau cach:").split()))
dic = {}
duong = []
am = []
zero = []
for i in lst:
    if i > 0:
        duong.append(i)
    elif i < 0:
        am.append(i)
    else:
        zero.append(i)
dic['positives'] = duong
dic['negatives'] = am
dic['zeros'] = zero
print(dic)
print("so duong:", len(duong), "so am:", len(am), "so 0:", len(zero))
```

```python
#W6A9
def nhap_dic():
    thu_vien = {}
    n = int(input("Nhap so cap trong dictionary"))
    for i in range(n):
        key = input("Nhap key")
        value = float(input("Nhap Value"))
        thu_vien[key] = value
    return thu_vien
dic1 = nhap_dic()
dic2 = nhap_dic()
dic = {}
diction = {}
for i in dic1:
    dic[i] = dic1[i]
    if i in dic2:
        dic[i] = dic1[i] + dic2[i]
for j in dic2:
    if j not in dic:
        dic[j] = dic2[j]
lst_key = sorted(dic)
for k in lst_key:
    diction[k] = dic[k]
print(diction)
```

```python
#W6A10
lst = list(map(int, input("Nhap cac so nguyen: ").split()))
k = int(input("Nhap so nguyen k: "))
danh_sach = []
for i in range(len(lst)):
    a = lst[i]
    for j in range(i + 1, len(lst)):
        b = lst[j]
        if a + b == k:
            tup = (a, b)
            rev = (b, a)
            if tup not in danh_sach and rev not in danh_sach:
                danh_sach.append(tup)
danh_sach = sorted(danh_sach)
print(danh_sach)
```

```python
#W6A12
lst = list(input().split())
m = int(lst[0])
repeat = 1
for i in range(len(lst)):
    count = lst.count(lst[i])
    if count > repeat:
        repeat = count
        m = int(lst[i])
    elif count == repeat:
        if m > int(lst[i]):
            m = int(lst[i])
print(m)
```
