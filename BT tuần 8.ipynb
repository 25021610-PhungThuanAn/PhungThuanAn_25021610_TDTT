```python
#W7A1
def binary_search(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        right = mid - 1
        return binary_search(arr, left, right, target)
    else:
        left = mid + 1
        return binary_search(arr, left, right, target)

arr = list(map(int, input().split()))
target = int(input())
left = 0
right = len(arr) - 1
print(binary_search(arr, left, right, target))
```

```python
#W7A3
def bubble_sort(arr):
    for i in range(len(arr) - 1):
        flag = False
        for j in range(len(arr) - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                flag = True
        if flag is False:
            break
    return arr

arr = list(map(int, input().strip().split()))
print(bubble_sort(arr))
```

```python
#W7A9
nums = eval(input())
m = int(input())
def binary_search(lst, aim):
    left = 0
    right = len(lst)
    while left < right:
        mid = (left + right) // 2
        if lst[mid] > aim:
            right = mid
        elif lst[mid] <= aim:
            left = mid + 1
    return left

tong = 0
max_mod = 0
prefix = []
for i in nums:
    tong = ( tong + i) % m
    max_mod = max(tong, max_mod)
    idx = binary_search(prefix, tong)
    if idx < len(prefix):
        so_lon_hon = prefix[idx]
        val = (tong - so_lon_hon + m) % m
        max_mod = max(max_mod, val)
    prefix.insert(idx, tong)

print(max_mod)
```

```python
#W7A11
day = list(map(int, input().split()))
lst = []
long = 0
dic = {}
for i in day:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
for _ in dic:
    num = _
    if (num + 1) in dic:
        if long < dic[num] + dic[num + 1]:
            long = dic[num] + dic[num + 1]
            best = num
for j in day:
    if j == best or j == best + 1:
        lst.append(j)


print(lst, long)
```
