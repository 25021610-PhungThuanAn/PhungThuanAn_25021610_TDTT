```python
#W6A28
day = list(map(int, input().split()))
count = []
for i in range(len(day) - 1):
    if day[i] > day[i + 1]:
        count.append(-1)
    elif day[i] < day[i + 1]:
        count.append(1)
    else:
        count.append(0)
if count.count(1) == len(count):
    print("tang thuc su")
elif count.count(-1) == len(count):
    print("giam thuc su")
else:
    print("vo danh")
```

```python
# So Amstrong
def so_amstrong(x):
    lst = []
    temp = x
    while temp > 0:
        so = temp % 10
        temp //= 10
        lst.append(so)
    mu = len(lst)
    tong = 0
    for i in lst:
        tong += i**(mu)
    if tong == x:
        return True
    return False
so_amstrong(1634)
```

```python
# So doi xung
def so_doi_xung(n):
    temp = n
    lst = []
    while temp > 0:
        so = temp % 10
        temp //= 10
        lst.append(so)
    for i in range(len(lst)):
        if lst[i] != lst[len(lst)-i-1]:
            return False
    return True
so_doi_xung(123321)
```
